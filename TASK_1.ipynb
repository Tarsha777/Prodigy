```python
df.drop(['Country/Territory','Capital','Continent','Area (km²)','Density (per km²)','World Population Percentage'],inplace=True,axis=1)
```

```python
df.drop(['Growth Rate'],inplace=True,axis=1)
```

```python
df.head()
```

```python
df_sorted=df.sort_values(by='2022 Population',ascending=False).head(10)
print(df_sorted)
```

```python
# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2022 Population", y="CCA3", data=df_sorted, palette="coolwarm")
plt.title("Top Ten Countries of Total Population (2022)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
# Create the bar plot
plt.figure(figsize=(15, 6))
df_sorted1=df.sort_values(by='2022 Population',ascending=True).head(10)
plt.subplot(2,2,1)
sns.barplot(x="2022 Population", y="CCA3", data=df_sorted1, palette="coolwarm")
plt.title("Top Ten Countries of Total Population (2022)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```
